The Nonlinear Least Squares (NLS) Regression Model

popt: Optimal values for the parameters

pcov:The estimated covariance of popt

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
def f(x, a, b, c):
    return a * np.exp(- b * x) + c
x = np.linspace(0, 4, 50)
y = f(x, a=2.5, b=1.3, c=0.5)
yi = y + 0.2 * np.random.normal(size=len(x))
```

```python
popt, pcov = curve_fit(f, x, yi)
a, b, c = popt
print(a, b, c)
yfitted = f(x, popt[0], popt[1], popt[2]) # or f(x,*pop)
plt.plot(x, yi, 'o', label='data ')
plt.plot(x, yfitted, '-', label='fit ')
plt.xlabel('x')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
x_data = np.linspace(-5, 5, num=50)
y_data = 2.9 * np.sin(1.5 * x_data) + np.random.normal(size=50)

def f(x, a, b):
    return a * np.sin(b * x)

params, params_covariance = optimize.curve_fit(f, x_data, y_data)
print(params)
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, f(x_data, params[0], params[1]),label='Fitted function')
plt.legend()
plt.show()
```
